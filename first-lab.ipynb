# Project SynthLab

**Objetivo Principal**: Criar uma plataforma de pesquisa de mercado e UX baseada em pessoas sintéticas (Synths) capazes de emular comportamento humano complexo.

## Gerar Synths Realistas

Os synths deve ser gerados atavés de um script que crie um JSON com atributos:

### Atributos Básicos
ex: id, nome, arquetipo, descrição, link photo, data de criação.
#### demográficos 
ex: Idade, Gênero Biológico, Identidade de gênero, Localização (País/Estado/Cidade), Escolaridade, Renda Mensal, Ocupação, Estado Civil, Composição Familiar.
#### psicográficos 
ex: Personalidade (Big Five), Valores, Interesses, Hobbies, Estilo de Vida, Atitudes, Motivações, Inclinação Política e Religiosa.
#### comportamentais realistas
ex: Hábitos de Consumo, Uso de Tecnologia, Padrões de Mídia, Fonte de Notícias, Comportamento de Compra, Lealdade à Marca, Engajamento em Redes Sociais.
#### limitações físicas e cognitivas
ex: Deficiências Visuais, Auditivas, Motoras, Cognitivas; N
#### capacidades tecnológicas
ex: Nível de Alfabetização Digital, Dispositivos Utilizados (novo, antigo), Zoom de fontes e Tela, velocidade de digitação, Frequência de Uso da Internet, Familiaridade com Aplicativos e Plataformas Online.
#### Lentes Comportamentais (Economia Comportamental)
ex: Aversão à Perda, Desconto Hiperbólico, Efeito Chamariz, Efeito de Ancoragem, Viés de Confirmação, Status Quo (Default Bias), Sobrecarga de Informação.

### Baseados em dados reais (IBGE, pesquisas de mercado, etc.)
Deve Garantir representação de diferentes grupos étnicos, socioeconômicos, idades, gêneros e localizações geográficas, muito em linha com a realidade Brasileira.

### Ferramentas Sugeridas
- Python para geração de dados
- Bibliotecas como Faker, NumPy, Pandas para manipulação de dados
- Modelos de Machine Learning para simular comportamentos complexos
